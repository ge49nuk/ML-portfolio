```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
cmap = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])
from Reg import Reg

X,y = datasets.make_regression(n_samples=1000, n_features=1, noise=10, random_state=4)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)

print("Data shape:",X_train.shape)
print("Example:",X_train[0])
print("Label shape:",y_train.shape)
print("Example:",y_train[1])

print("\nVisualization: Data distribution")
fig = plt.figure(figsize=(8,6))
plt.scatter(X[:,0], y, color = "b", marker="o", s = 30)
plt.show()
```

```python
reg = Reg(it=1050,lr=0.01)
reg.fit(X_train, y_train)
ypred = reg.predict(X_test)

mse = 0
for i in range(len(X_test)):
    mse+=pow(y_test[i]-ypred[i],2)
mse/=len(X_test)
print("MSE loss", mse[0])

print("\nVisualization: Prediction")
fig = plt.figure(figsize=(8,6))
plt.scatter(X_test[:,0], y_test, color = "b", marker="o", s = 30)
plt.scatter(X_test[:,0], ypred, color = "r", marker="o", s = 30)
plt.show()
```
