```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
cmap = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])
from logreg import LogReg

bc = datasets.load_breast_cancer()
X,y = bc.data, bc.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)

print("Data shape:",X_train.shape)
print("Example:",X_train[0])
print("Label shape:",y_train.shape)
print("Example:",y_train[1])

print("\nVisualization: Data distribution")
fig = plt.figure(figsize=(8,6))
plt.scatter(X[:,0], y, color = "b", marker="o", s = 30)
plt.show()
```

```python
logr = LogReg(lr= 0.0001, it = 5000)
logr.fit(X_train, y_train)
y_pred = logr.predict(X_test)
print("Accuracy:", np.sum(y_pred==y_test)/len(y_test))

print("\nVisualization: Prediction")
fig = plt.figure(figsize=(8,6))
plt.scatter(X_test[:,0], y_pred, color = "b", marker="o", s = 50)
plt.scatter(X_test[:,0], y_test, color = "r", marker="o", s = 50)
plt.show()
```
